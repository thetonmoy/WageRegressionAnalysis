```python
import pandas as pd
```

```python
data = pd.read_csv("DT-Wage.csv")
```

```python
category_counts = data['region'].value_counts()
print(category_counts)
```

```python
data = data.drop('region', axis=1)
data
```

```python
category_counts = data['maritl'].value_counts()
print(category_counts)
```

```python
data['maritl'] = data['maritl'].str.replace(r'^\d+\.\s*', '', regex=True)
data
```

```python
category_counts = data['maritl'].value_counts()
print(category_counts)
```

```python
data = pd.get_dummies(data, columns=['maritl'], prefix='maritl')
data
```
